```python
#import nglview as nv
import numpy as np
from numpy import array
import os
import glob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm
import matplotlib.ticker as ticker
%matplotlib inline
import tqdm
plt.style.use('ggplot')
```

```python
def runningMeanFast(x,N):
    return np.convolve(x,np.ones((N,))/N)[(N-1):]
```

```python
dataDir='/home/wjuan/lipid_kc'
filepath='/'.join([dataDir,'popc_avg_press.txt'])
dataPC=pd.read_table(filepath,header=None,delim_whitespace=True)
dataPC.columns=['Z','p_avg']
dataPC['run_avg']=runningMeanFast(dataPC['p_avg'],3)
dataPC['rZ']=runningMeanFast(dataPC['Z'],3)
print (np.shape(dataPC))
#print (dataPC.head())
filepath='/'.join([dataDir,'popc_chunk1_press.txt'])
dPC1=pd.read_table(filepath,header=None,delim_whitespace=True)
dPC1.columns=['z','press']
#print (dPC1.head())
filepath='/'.join([dataDir,'popc_chunk2_press.txt'])
dPC2=pd.read_table(filepath,header=None,delim_whitespace=True)
dPC2.columns=['z','press']
#print (dPC2.head())
filepath='/'.join([dataDir,'popc_chunk3_press.txt'])
dPC3=pd.read_table(filepath,header=None,delim_whitespace=True)
dPC3.columns=['z','press']
#print (dPC3.head())
filepath='/'.join([dataDir,'popc_chunk4_press.txt'])
dPC4=pd.read_table(filepath,header=None,delim_whitespace=True)
dPC4.columns=['z','press']
#print (dPC4.head())
dataPC['c1']=dPC1['press']
dataPC['c2']=dPC2['press']
dataPC['c3']=dPC3['press']
dataPC['c4']=dPC4['press']
dataPC['std']=dataPC[['c1','c2','c3','c4']].std(axis=1)
dataPC['cmean']=dataPC[['c1','c2','c3','c4']].mean(axis=1)
dataPC.head()
```
