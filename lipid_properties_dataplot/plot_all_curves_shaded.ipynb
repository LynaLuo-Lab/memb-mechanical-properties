```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.style.use('ggplot')
```

```python
!pwd
def runningMeanFast(x,N):
    return np.convolve(x,np.ones((N,))/N)[(N-1):]
```

```python
##################  charge density PIP2 #########################

dataPI_c3=pd.read_table('lipid_kc/charge_PIP2.txt',header=None,delim_whitespace=True)
dataPI_c3.columns=['nX','nY','nZ','Phosphate_P']
groupedPI_c3=dataPI_c3['Phosphate_P'].groupby(dataPI_c3['nZ'])
#print (groupedPI_c3.head())

dataPC_c3=pd.read_table('lipid_kc/charge_POPC.txt',header=None,delim_whitespace=True)
dataPC_c3.columns=['nX','nY','nZ','Phosphate_P']
groupedPC_c3=dataPC_c3['Phosphate_P'].groupby(dataPC_c3['nZ'])
#print (groupedPI_c3.head())


dataPS_c3=pd.read_table('lipid_kc/charge_POPS.txt',header=None,delim_whitespace=True)
dataPS_c3.columns=['nX','nY','nZ','Phosphate_P']
groupedPS_c3=dataPS_c3['Phosphate_P'].groupby(dataPS_c3['nZ'])
#print (groupedPI_c3.head())

dataCHO_c3=pd.read_table('lipid_kc/charge_CHOL.txt',header=None,delim_whitespace=True)
dataCHO_c3.columns=['nX','nY','nZ','Phosphate_P']
groupedCHO_c3=dataCHO_c3['Phosphate_P'].groupby(dataCHO_c3['nZ'])
#print (groupedPI_c3.head())

##################  Mass density CHOL #########################

dataCH_c3=pd.read_table('lipid_kc/volmap_CHOL_O3.txt',header=None,delim_whitespace=True)
dataCH_c3.columns=['nX','nY','nZ','Phosphate_P']
groupedCH_c3=dataCH_c3['Phosphate_P'].groupby(dataCH_c3['nZ'])
#print (groupedCH_c3.head())


dataCH_ali=pd.read_table('lipid_kc/volmap_CHOL_C25.txt',header=None,delim_whitespace=True)
dataCH_ali.columns=['nX','nY','nZ','Aliphatic_tails']
groupedCH_ali=dataCH_ali['Aliphatic_tails'].groupby(dataCH_ali['nZ'])


##################  Mass density POPC  #########################

dataHS2=pd.read_table('lipid_kc/volmap_POPC_P.txt',header=None,delim_whitespace=True)
dataHS2.columns=['nX','nY','nZ','Phosphate_P']
groupedHS2=dataHS2['Phosphate_P'].groupby(dataHS2['nZ'])
#print (groupedHS2.head())

dataN2=pd.read_table('lipid_kc/volmap_POPC_ester.txt',header=None,delim_whitespace=True)
dataN2.columns=['nX','nY','nZ','Ester']
groupedN2=dataN2['Ester'].groupby(dataN2['nZ'])

dataC2=pd.read_table('lipid_kc/volmap_POPC_alipha.txt',header=None,delim_whitespace=True)
dataC2.columns=['nX','nY','nZ','Aliphatic_tails']
groupedC2=dataC2['Aliphatic_tails'].groupby(dataC2['nZ'])

dataHSS=pd.read_table('lipid_kc/volmap_POPC_glyc.txt',header=None,delim_whitespace=True)
dataHSS.columns=['nX','nY','nZ','Glycerol_linker']
groupedHSS=dataHSS['Glycerol_linker'].groupby(dataHSS['nZ'])

dataHSSs=pd.read_table('lipid_kc/volmap_POPC_teramine.txt',header=None,delim_whitespace=True)
dataHSSs.columns=['nX','nY','nZ','TertiaryAmine_group']
groupedHSSs=dataHSSs['TertiaryAmine_group'].groupby(dataHSSs['nZ'])

##################  Mass density PIP2  #########################

dataHSii=pd.read_table('lipid_kc/volmap_PIP2_P.txt',header=None,delim_whitespace=True)
dataHSii.columns=['nX','nY','nZ','Phosphate_P']
groupedHSii=dataHSii['Phosphate_P'].groupby(dataHSii['nZ'])

dataHSi=pd.read_table('lipid_kc/volmap_PIP2_P4_P5.txt',header=None,delim_whitespace=True)
dataHSi.columns=['nX','nY','nZ','Phosphate_P4_P5']
groupedHSi=dataHSi['Phosphate_P4_P5'].groupby(dataHSi['nZ'])


dataNi=pd.read_table('lipid_kc/volmap_PIP2_ester.txt',header=None,delim_whitespace=True)
dataNi.columns=['nX','nY','nZ','Ester']
groupedNi=dataNi['Ester'].groupby(dataNi['nZ'])

dataCi=pd.read_table('lipid_kc/volmap_PIP2_alipha.txt',header=None,delim_whitespace=True)
dataCi.columns=['nX','nY','nZ','Aliphatic_tails']
groupedCi=dataCi['Aliphatic_tails'].groupby(dataCi['nZ'])

dataCCi=pd.read_table('lipid_kc/volmap_PIP2_glyc.txt',header=None,delim_whitespace=True)
dataCCi.columns=['nX','nY','nZ','Glycerol_linker']
groupedCCi=dataCCi['Glycerol_linker'].groupby(dataCCi['nZ'])


##################  Mass density POPS  #########################

dataPS_P=pd.read_table('lipid_kc/volmap_POPS_P.txt',header=None,delim_whitespace=True)
dataPS_P.columns=['nX','nY','nZ','Phosphate_P']
groupedPS_P=dataPS_P['Phosphate_P'].groupby(dataPS_P['nZ'])

dataPS_COO=pd.read_table('lipid_kc/volmap_POPS_COO.txt',header=None,delim_whitespace=True)
dataPS_COO.columns=['nX','nY','nZ','Phosphate_P4_P5']
groupedPS_COO=dataPS_COO['Phosphate_P4_P5'].groupby(dataPS_COO['nZ'])


dataPS_ester=pd.read_table('lipid_kc/volmap_POPS_ester.txt',header=None,delim_whitespace=True)
dataPS_ester.columns=['nX','nY','nZ','Ester']
groupedPS_ester=dataPS_ester['Ester'].groupby(dataPS_ester['nZ'])

dataPS_ali=pd.read_table('lipid_kc/volmap_POPS_alipha.txt',header=None,delim_whitespace=True)
dataPS_ali.columns=['nX','nY','nZ','Aliphatic_tails']
groupedPS_ali=dataPS_ali['Aliphatic_tails'].groupby(dataPS_ali['nZ'])

dataPS_glyc=pd.read_table('lipid_kc/volmap_POPS_glyc.txt',header=None,delim_whitespace=True)
dataPS_glyc.columns=['nX','nY','nZ','Glycerol_linker']
groupedPS_glyc=dataPS_glyc['Glycerol_linker'].groupby(dataPS_glyc['nZ'])

```

```python
plt.rcParams['axes.facecolor']='white'
import matplotlib
matplotlib.rc('axes',edgecolor='k')
fig,(ax1,ax2,ax3,ax4,ax5,ax6) = plt.subplots(6,1,sharex=True,figsize=(16,30))


ax1.plot(dataPC['Z'],  dataPC['p_avg'],    color='g', linewidth=3, label='PC:PC')
ax1.fill_between(dataPC['Z'],dataPC['p_avg']-dataPC['std'],dataPC['p_avg']+dataPC['std'],color='g', alpha=0.3)
ax1.plot(dataPS['Z'],  dataPS['p_avg'],    color='b', linewidth=3, label='PC:PS')
ax1.fill_between(dataPS['Z'],dataPS['p_avg']-dataPS['std'],dataPS['p_avg']+dataPS['std'],color='b', alpha=0.3)
ax1.plot(dataPI['Z'],  dataPI['p_avg'],    color='r', linewidth=3, label='PC:PIP2')
ax1.fill_between(dataPI['Z'],dataPI['p_avg']-dataPI['std'],dataPI['p_avg']+dataPI['std'],color='r', alpha=0.3)
ax1.plot(dataCHOL['Z'],dataCHOL['p_avg'],  color='gray', linewidth=3, label='PC:Chol:PIP2')
ax1.fill_between(dataCHOL['Z'],dataCHOL['p_avg']-dataCHOL['std'],dataCHOL['p_avg']+dataCHOL['std'],color='gray', alpha=0.3)
ax1.set_ylabel(r"Pressure $P$ (Bar)",color="black",fontsize=22)
ax1.tick_params(axis='x', colors='black')
ax1.tick_params(axis='y', colors='black')
ax1.legend(loc='center left',bbox_to_anchor=(1,0.5),fontsize=25)
ax1.tick_params(axis='y', which='major', labelsize=25)
ax1.set_title('Lateral pressure profile',fontsize=25)

ax2.plot(groupedPC_c3.mean(),   linewidth=4, label='PC',color='red')
ax2.plot(groupedPS_c3.mean(),   linewidth=4, label='PS',color='blue') 
ax2.plot(groupedPI_c3.mean(),   linewidth=4, label='PIP2',color='cyan') 
ax2.plot(groupedCHO_c3.mean(),   linewidth=4, label='Chol',color='green') 
ax2.set_title('Charge Distribution',fontsize=25)
ax2.set_ylabel(r"Charge Density ($e/m^3$)",color="black",fontsize=22)
ax2.tick_params(axis='x', colors='black')
ax2.tick_params(axis='y', colors='black')
ax2.legend(loc='center left',bbox_to_anchor=(1,0.5),fontsize=25)
ax2.tick_params(axis='y', which='major', labelsize=25)

ax3.plot(groupedCH_c3.mean(),   linewidth=4, label='Chol-O3',color='red')
ax3.plot(groupedCH_ali.mean(),  linewidth=4, label='Chol-C25',color='blue')
ax3.set_title('CHOL',fontsize=25)
ax3.set_ylabel(r"Mass Density ($Da/m^3$)",color="black",fontsize=22)
ax3.tick_params(axis='x', colors='black')
ax3.tick_params(axis='y', colors='black')
ax3.legend(loc='center left',bbox_to_anchor=(1,0.5),fontsize=25)
ax3.tick_params(axis='y', which='major', labelsize=25)

ax4.plot(groupedHS2.mean(), linewidth=4, label='PC-P',color='red')
ax4.plot(groupedHSSs.mean(),linewidth=4, label='PC-TertiaryAmine',color='cyan')
ax4.plot(groupedN2.mean(),  linewidth=4, label='PC-Ester',color='green')
ax4.plot(groupedC2.mean(),  linewidth=4, label='PC-AliphaticTails',color='blue')
ax4.plot(groupedHSS.mean(), linewidth=4, label='PC-Glycerol',color='black')
ax4.set_title('POPC',fontsize=25)
ax4.set_ylabel(r"Mass Density ($Da/m^3$)",color="black",fontsize=22)
ax4.tick_params(axis='x', colors='black')
ax4.tick_params(axis='y', colors='black')
ax4.legend(loc='center left',bbox_to_anchor=(1,0.5),fontsize=25)
ax4.tick_params(axis='y', which='major', labelsize=25)

ax5.plot(groupedHSii.mean(),linewidth=4, label='PIP2-P', color='red') # positive without '-' negative sign, negative, should consider '-' sign
ax5.plot(groupedHSi.mean(), linewidth=4, label='PIP2-P4_P5', color='magenta') # positive without '-' negative sign, negative, should consider '-' sign
ax5.plot(groupedNi.mean(),  linewidth=4, label='PIP2-Ester', color='green')
ax5.plot(groupedCi.mean(),  linewidth=4, label='PIP2-AliphaticTails', color='blue')
ax5.plot(groupedCCi.mean(), linewidth=4, label='PIP2-Glycerol', color='black')
ax5.set_title('PIP2',fontsize=25)
ax5.set_ylabel(r"Mass Density ($Da/m^3$)",color="black",fontsize=22)
ax5.tick_params(axis='x', colors='black')
ax5.tick_params(axis='y', colors='black')
ax5.tick_params(axis='x',which='major',labelsize=25)
ax5.tick_params(axis='y',which='major',labelsize=25)
ax5.legend(loc='center left',bbox_to_anchor=(1,0.5),fontsize=25)
#ax5.set_xlabel(r"Z-axis ($\AA$)",color="black",fontsize=35)

ax6.plot(groupedPS_P.mean(),linewidth=4, label='PS-P', color='red') # positive without '-' negative sign, negative, should consider '-' sign
ax6.plot(groupedPS_COO.mean(), linewidth=4, label='PS-COO', color='magenta') # positive without '-' negative sign, negative, should consider '-' sign
ax6.plot(groupedPS_ester.mean(),  linewidth=4, label='PS-Ester', color='green')
ax6.plot(groupedPS_ali.mean(),  linewidth=4, label='PS-AliphaticTails', color='blue')
ax6.plot(groupedPS_glyc.mean(), linewidth=4, label='PS-Glycerol', color='black')
ax6.set_title('POPS',fontsize=25)
ax6.set_ylabel(r"Mass Density ($Da/m^3$)",color="black",fontsize=22)
ax6.tick_params(axis='x', colors='black')
ax6.tick_params(axis='y', colors='black')
ax6.tick_params(axis='x',which='major',labelsize=25)
ax6.tick_params(axis='y',which='major',labelsize=25)
ax6.legend(loc='center left',bbox_to_anchor=(1,0.5),fontsize=25)
ax6.set_xlabel(r"Z-axis ($\AA$)",color="black",fontsize=35)
ax6.set_xlim(-50,50)
ax6.set_xticks(np.arange(-50,60,10))


plt.show()
```
